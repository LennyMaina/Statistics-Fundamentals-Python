```python
from math import sqrt
```

```python
sample_size = 40
satisfied = 32
```

```python
sample_proportion = satisfied / sample_size
z_score = 1.96  # for 95% confidence level
```

```python
standard_error = sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)
standard_error
```

```python
lower = sample_proportion - z_score * standard_error
upper = sample_proportion + z_score * standard_error
```

```python
print('sample_proportion:', sample_proportion)
print('lower bound:', lower)
print('upper bound:', upper)
```
